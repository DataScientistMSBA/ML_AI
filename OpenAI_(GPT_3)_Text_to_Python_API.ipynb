OpenAI Text-to-Python API

```python
def RunAPI():
  prompt = f'''
  """
  {input()}
  """
  '''

  response = openai.Completion.create(
    model="code-davinci-002",
    prompt=prompt,
    temperature=0,
    max_tokens=4000,
    top_p=1,
    frequency_penalty=0,
    presence_penalty=0
  )

  output = response['choices'][0]['text']

  lim=75
  for s in output.split("\n"):
      if s == "": print
      w=0 
      l = []
      for d in s.split():
          if w + len(d) + 1 <= lim:
              l.append(d)
              w += len(d) + 1 
          else:
              print(" ".join(l))
              l = [d] 
              w = len(d)
      if (len(l)): print(" ".join(l))
```

```python
RunAPI()
```
